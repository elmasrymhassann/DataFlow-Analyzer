### <span style="color:orange"><u>**18. Sort a tuple by its float element**<u></span>

```python
#from hashlib import new


items = (("item1", 10.5), ("item2", 5.7), ("item3", 8.9))

# TODO: sort by the float value
# your code here
new_item = list(items)
for i in range(len(new_item)):
    for j in range(i + 1, len(new_item)):
        if new_item[i][1] > new_item[j][1]:
            new_item[i], new_item[j] = new_item[j], new_item[i]

print(new_item)
#print(sorted_items)
# Hint: Use sorted() with a key function.
```
