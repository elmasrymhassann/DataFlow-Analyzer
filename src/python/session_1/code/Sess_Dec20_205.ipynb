```python
for i in range(6): 
    if i == 3:
        continue # continue will continue the loop in all case of if true in contrast of "break"
    print(i)
```

```python
for i in range(6): 
    if i == 3:
        break # break will stop the loop if the condition is true
    print(i)
```

```python
for i in range(6): 
    if i == 3:
        print ("if you see this message it mean the condition 'pass' happend")
        pass # pass will do nothing and continue the loop
    print(i)
```
